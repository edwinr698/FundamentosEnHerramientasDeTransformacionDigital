**ACTIVIDA**

El dataset de **Black Friday** contiene la información sobre los compradores del viernes negro de 2005 a 2013. Los datos se almacenaron en función del ID del producto, es decir, si una persona compró 10 productos, sus datos se almacenaron en 10 registros.

Utilice la librería Pandas para extraer información de dataset y poder analizar información de los productos, los compradores y la relación entre ellos.

```
# Importar librerías
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Leer el archivo .CSV
# Leer los datos desde el GitHub

df = pd.read_csv(
    "https://raw.githubusercontent.com/jaguzman/dataSets/master/datasets/BlackFriday.csv",
    sep = ',',         # Se especifica el separador de los datos
    thousands = None,  # Se especifica el separador de miles
    decimal = '.')     # Se especifica el separador de los decimales
```

```
# Mostrar los primeros datos del dataframe
df.__()
```

```
# Obtener los nombres de las columnas
df.__
```

```
# Obtener las dimensiones, similar a numpy
df.__
```

```
# Imprimir información general, como el total de columnas y filas, y sus tipos
df.__()
```

```
# seleccionar rangos utilizando los nombres de las columnas de 'Product_Category_1' hasta 'Purchase'
df.loc[:, 'Product_Category_1':'__']
```

```
# Ordenar según la columna 'Purchase'
df.sort_values(by="__")
```

```
# Eliminar las columnas 'User_ID' y 'Product_ID'
df = df.drop(columns=['__', '__'])
df
```

```
# Para filtrar las filas que cumplan cierta condición: comprado por una mujer
df[df['Gender'] == '_']
```

```
# Grafiar el número de compras por el rando de edades

# Tamaño de la gráfica
plt.figure(figsize=(10,4))
# Gráficar
df["__"].value_counts().plot(kind='bar')
# explode: típo de gráfica (barra)
```

```
# Ahora graficaremos la cantidad mujeres y de hombres que realizaron compras

# Tamaño de la gráfica
plt.figure(figsize=(5,5)) 
# Título de la imagen y tamaño del título
plt.title('Género', fontsize=20) 
# Gráficar
df["__"].value_counts().plot.pie(explode=[0,0.25], autopct='%1.1f%%', shadow=True, colors=["#33FFB0", "#33A1FF"], fontsize=14, startangle=25)
```
